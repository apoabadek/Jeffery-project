```python
import seaborn as sns
import matplotlib.pyplot as plt
numerical_columns = redData.select_dtypes(include='number').columns
num_columns = len(numerical_columns)
rows = (num_columns // 3) + (num_columns % 3)
cols = 3
plt.figure(figsize=(15,10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(rows, cols, i)
    sns.violinplot(x=column, data=redData, color='red')
    plt.title(f'Violin Plot of {column}')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
numerical_columns = redData.select_dtypes(include='number').columns
num_columns = len(numerical_columns)
rows = (num_columns // 3) + (num_columns % 3)
cols = 3
plt.figure(figsize=(15,10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(rows, cols, i)
    sns.countplot(x=column, data=redData, color='skyblue')
    plt.title(f'Count Plot of {column}')
plt.tight_layout()
plt.show()
```
